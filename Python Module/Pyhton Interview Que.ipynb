```python
lst_1 = [1, 2, 3, 4]
lst_2 = lst_1.copy()
lst_1[0] = 100
lst_2[0] = 200
print(lst_1)
print(lst_2) 
```

```python
lst_1 = [[1, 2], [3, 4]]
lst_2 = lst_1.copy()
lst_1[0][0] = 100 #nested list mai chahe hum lst1 m ya lst2 m change kren dono m reflect hoga eska sol h deepcopy 
print(lst_1)
print(lst_2) 
```

```python
import copy     
lst_1 = [[1, 2], [3, 4]]
lst_2 = copy.deepcopy(lst_1)
lst_1[0][0] = 100 
lst_2[0][0] = 200
print(lst_1)
print(lst_2)
```

```python
#other way of calculating frequency without using inbuilt fun
s = "aabbccdd"
freq = {}
for char in s:
    freq[char] = freq.get(char,0)+1
print(freq)
```

```python
#factorial
def fact(a):
    if a == 0:
        return 1
    else:
        return a * fact(a-1)
    
print(fact(5))
```
